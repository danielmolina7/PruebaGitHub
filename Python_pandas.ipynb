```python
#se importa pandas
import pandas as pd
#se crea el dataframe. como en la hoja hay caracteres que no abarca la codificacion utf-8 debe usarse una codificacion alternativa

df=pd.read_csv("medicamentos_colombia.csv/medicamentos_colombia.csv", encoding="ISO-8859-1")

df.head()
```

```python
# 1. filtrar por laboratorio

fabricante=df[df["Fabricante"]=="Pfizer"]
print(fabricante)
```

```python
# 2. Buscar medicamento especifico

print(df[df["Nombre del Medicamento"]=="Loratadina"])

```

```python
# 3. estadistica de precios

print(df.describe())
```

```python
# 4. agrupar datos

grouped = df.groupby('Nombre del Medicamento')['Uso'].apply(list)
print(grouped)
```

```python
# 5. Ordenar datos

df_sorted= df.sort_values(by="Precio (COP)", ascending=True)
print(df_sorted)
```

```python
# 6. Filtrar medicamentos por Laboratorio y precio
df_lab_precio = df[(df['Fabricante']== 'Genfar') & (df['Precio (COP)'] > 5000)]

# Mostrar las primeras filas del filtro
print(df_lab_precio.head())
```

```python
# 7. Añadir función de descuento con Apply

def aplicar_descuento(precio):
    return precio * 0.9

df['Precio_descuento'] = df['Precio (COP)'].apply(aplicar_descuento)

print(df.head())

```

```python
# 8. Manejo de tipos de datos

df['Precio (COP)'] = df['Precio (COP)'].astype(float)
print(df.dtypes)
```

```python

# 9. Eliminar colummnas

df = df.drop ('Precio (COP)', axis=1)

print(df)

```

```python
# 10. Promedio de precio por laboratorio
#recordar volver a cargar el dataframe puesto que en el #9. se eliminó la columna de Precio (COP)

promedio_precio = df.groupby("Fabricante") ["Precio (COP)"].mean().reset_index()

promedio_precio.rename(columns={"Precio (COP)": 'Promedio_Precio'}, inplace=True)

print(promedio_precio)


```
